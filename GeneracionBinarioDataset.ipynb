# Imports

```python
import glob
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

# Carga de los datos

```python
def load_dataset(path):
    cont = 0
    X = []

    for img in glob.glob(path + '*.png'):
        if cont % 100 == 0:
            print(str(cont) + ' imagenes cargadas')
        cont = cont + 1

        X.append(plt.imread(img).astype('float32'))

    return np.array(X)
```

```python
X = load_dataset(FILE_PATH)
```

```python
with open('binary_60x80.pkl', 'wb') as file:
    pickle.dump(X, file)
```
